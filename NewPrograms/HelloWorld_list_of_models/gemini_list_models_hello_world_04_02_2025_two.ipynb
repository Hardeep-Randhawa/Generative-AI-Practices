```python
from google import genai
from google.colab import userdata
import json

#API_KEY = put the new api key here created latest
api_key = userdata.get("GOOGLE_API_KEY")
client = genai.Client(api_key=api_key)

all_models = list(client.models.list())

# sanity check: list models
for m in all_models[-10:]:
    model_info=client.models.get(model=m.name)
    print("Model Name",{model_info.name},"Description - ",{model_info.description})

# generate content
#response = client.models.generate_content(
#    model="gemini-1.5-flash",
#    contents="What is the meaning of life?")

#print(response.text)
```
